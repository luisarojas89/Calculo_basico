#Función seccionada


```python
def H(X):
  Y = np.zeros(len(X))
  for idx,x in enumerate(X):
    if x>=0:
      Y[idx] = 1.0
  return Y

# Datos para graficación

N = 1000

X = np.linspace(-1,1, num=N)

y = H(X)

fig, ax = plt.subplots()
ax.plot(X, y)
ax.grid()
```

```python
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
```

#Gráfica en 3D

```python
def f(x,y): #Función
  return np.sin(x) + 2*np.cos(y)

#Vectores
res=100
x=np.linspace(-4,4,res)
y=np.linspace(-4,4,res)

#Plano generado
x,y=np.meshgrid(x,y)

z=f(x,y)

#Gráficas
fig,ax = plt.subplots(subplot_kw={"projection":"3d"})
surf=ax.plot_surface(x,y,z, cmap=cm.cool)
fig.colorbar(surf)
```

#Curvas de nivel

```python
fig2, ax2 = plt.subplots()
level_map= np.linspace(np.min(z), np.max(z), num=10)

#Graficar curvas de nivel
cp = ax2.contourf(x,y,z,levels=level_map,cmap=cm.cool)
fig2.colorbar(cp)
```

# **Descenso del gradiente**

#Dependencias

```python
from matplotlib import cm #Para mejorar colores
import numpy as np
import matplotlib.pyplot as plt
```

### Gráfica en 3D de nuestra función de coste

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x,y):
  return x**2 + y**2;

#Crear vectores
res = 100
X = np.linspace(-4, 4, res)
Y = np.linspace(-4, 4, res)

X, Y = np.meshgrid(X, Y)

Z = f(X,Y) 

# Gráficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap=cm.gist_ncar,
                       linewidth=0, antialiased=False)
fig.colorbar(surf)
```

## Descenso del gradiente


```python
level_map= np.linspace(np.min(Z),np.max(Z),res)
plt.contourf(X,Y,Z,level_map, cmap=cm.gist_ncar)
plt.colorbar()
plt.title("Descenso del gradiente")

#Algortimo de optimización:
p=np.random.rand(2)*8 - 4 #Genera número aleatorios entre 0 y 1, genera 2.
plt.plot(p[0],p[1], "o", c="k") #Graficar punto

h=0.01
lr=0.01

def derivate(_p,p):
  return  (f(_p[0],_p[1]) - f(p[0],p[1])) / h

def gradient(p):
    grad = np.zeros(2)
    for idx, val in enumerate(p):
        cp = np.copy(p)
        cp[idx]= cp[idx]+h

        dp=derivate(cp,p)
        grad[idx] = dp
    return grad

for i in range(100):
    p=p -lr *gradient(p)
    if (i % 10 ==0):
      plt.plot(p[0],p[1],"o",c="r")

plt.plot(p[0],p[1],"o",c="w")
plt.show()
  

```
